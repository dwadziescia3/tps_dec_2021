```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/train.csv', index_col='Id')
```

```python
df.head()
```

```python
df.columns
```

```python
words = ['Soil', 'Wilderness']

categorical = [any([word in col for word in words]) for col in df.columns.tolist()]
print(categorical)
```

```python
df.loc[:, [not elem for elem in categorical]].corr()
```

```python
df.loc[:, categorical].corr()
```

```python
numerical = [not elem for elem in categorical]
```

```python
from sklearn.feature_selection import f_classif
```

```python
f_values = f_classif(df.loc[:, numerical].drop('Cover_Type', axis=1), df['Cover_Type'])
```

```python
f_values
```

```python
f_values[1]
```

```python
df.columns[numerical]
```

```python
p_values = pd.Series(index=df.loc[:, numerical].drop('Cover_Type', axis=1).columns, data=f_values[1])
p_values
```

```python
threshold = 0.05
p_values < threshold
```

```python
from sklearn.feature_selection import mutual_info_classif
```

```python
mi = mutual_info_classif(df.loc[:, categorical], df['Cover_Type'], discrete_features=True)
```

```python
mi_series = pd.Series(data = mi, index=df.loc[:, categorical].columns)
mi_series.sort_values(ascending=False).head(50)
```
